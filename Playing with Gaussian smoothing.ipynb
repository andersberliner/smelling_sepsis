```python
import numpy as np
from scipy.ndimage.filters import gaussian_filter1d
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0,10,0.5)
x = np.sin(t) + np.random.randn(len(t))*0.33

xT = x.T
plt.figure(figsize=(20,10))
plt.plot(t, x,'bo-',label='1xN', linewidth=5, alpha=0.5)
plt.plot(t.T, x.T, 'rd-', label='Nx1')
plt.plot(t,gaussian_filter1d(x,2), 'b-', label='Smoothed 1xN', linewidth=10, alpha=0.25)
plt.plot(t,gaussian_filter1d(xT, 2), 'r-' ,label='smoothed Nx1')
plt.legend()

plt.figure(figsize=(20,10))
plt.plot(t,x, label='raw', linewidth=3, alpha=0.5)
plt.plot(t,np.sin(t), 'k--', label='exact')

for sigma in [0.5, 1, 2]:
    y = gaussian_filter1d(x, sigma)
    plt.plot(t,y, label='/sigma = %s'%str(sigma))
    
plt.legend(loc=0)
```
